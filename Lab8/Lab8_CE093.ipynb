```python
y = df[["v1"]].to_numpy().flatten()
print(y)
```

```python
le = LabelEncoder()
y = le.fit_transform(y)
print(y)
```

# **Naive Bayes**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.naive_bayes import GaussianNB, MultinomialNB
from sklearn.metrics import precision_score, recall_score, accuracy_score
```

```python
model1 = MultinomialNB()
model1.fit(X_train, y_train)
y_pred1 = model1.predict(X_test)

print('Accuracy: %.3f' % accuracy_score(y_test, y_pred1))
print('Precision: %.3f' % precision_score(y_test, y_pred1))
print('Recall: %.3f' % recall_score(y_test, y_pred1))
```

#  Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dtc1 = DecisionTreeClassifier(criterion = 'entropy', random_state = 69)
dtc1.fit(X_train, y_train)
pred1 = dtc1.predict(X_test)

print('Accuracy: %.3f' % accuracy_score(y_test, pred1))
print('Precision: %.3f' % precision_score(y_test, pred1))
print('Recall: %.3f' % recall_score(y_test, pred1))
```
